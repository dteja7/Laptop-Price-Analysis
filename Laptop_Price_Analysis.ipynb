```python
def operatingsystem(x):
    if x=='Windows 10' or x=='Windows 10 S' or x=='Windows 7':
     return 'Windows'
    elif x=='macOS' or x=='Mac OS X':
     return 'Mac'
    elif x=='Linux':
     return 'Linux'
    elif x=='No OS':
     return 'No OS'
    else:
     return 'Others'
df['os']=df['OS'].apply(operatingsystem)
df.drop(columns=['OS'],inplace= True)
```

Model Selection

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
!pip install category_encoders
import category_encoders as ce
!pip install scikit-learn
from sklearn.impute import SimpleImputer
```

```python
X = df.drop('Price_euros', axis=1)
y = np.log(df['Price_euros'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.15, random_state=42)
X_train_encoded = X_train.copy()
X_test_encoded = X_test.copy()

# Convert 'Product' to an index and then one-hot encode
for df in [X_train_encoded, X_test_encoded]:
    df['Product'] = df['Product'].astype('category').cat.codes  # Convert to category codes
```

```python
X.head()
```

```python
categorical_features = ['Company','Product', 'TypeName', 'CPU_company', 'GPU_company',
                        'os', 'CPU_model', 'Screen', 'IPSpanel','GPU_model',
                         'PrimaryStorageType','SecondaryStorageType']

# Create a OneHotEncoder object
encoder = ce.OneHotEncoder(cols=categorical_features, handle_unknown='ignore')

# Fit the encoder on the training data and transform both train and test data
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)

# Map 'RetinaDisplay' and 'Touchscreen' to numerical values before fitting the model
for df in [X_train_encoded, X_test_encoded]:
    df['RetinaDisplay'] = df['RetinaDisplay'].map({'Yes': 1, 'No': 0})
    df['Touchscreen'] = df['Touchscreen'].map({'Yes': 1, 'No': 0})

# Now, we can fit our Linear Regression model
model = LinearRegression()
model.fit(X_train_encoded, y_train)
# Create a SimpleImputer to replace NaN with the most frequent value
imputer = SimpleImputer(strategy='most_frequent')

# Fit the imputer on the training data and transform both train and test data
X_train_encoded = imputer.fit_transform(X_train_encoded)
X_test_encoded = imputer.transform(X_test_encoded)
```

```python
y_pred = model.predict(X_test_encoded)
r_squared = r2_score(y_test, y_pred)
print("R-squared:",r_squared)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

```python
# Now, we can fit our Random Forest model
model = RandomForestRegressor()
model.fit(X_train_encoded, y_train)
# Create a SimpleImputer to replace NaN with the most frequent value
imputer = SimpleImputer(strategy='most_frequent')

# Fit the imputer on the training data and transform both train and test data
X_train_encoded = imputer.fit_transform(X_train_encoded)
X_test_encoded = imputer.transform(X_test_encoded)
```

```python
# Now, we can fit our Random Forest model
model = RandomForestRegressor()
model.fit(X_train_encoded, y_train)
y_pred = model.predict(X_test_encoded)
r_squared = r2_score(y_test, y_pred)
print("R-squared:",r_squared)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```
